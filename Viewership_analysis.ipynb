## Aggregate data by time

### Subtask:
Group the data by date and calculate the total time watched for each day.


```python
#group data by date and calculate the total time watched for each day
```

**Reasoning**:
Convert the 'TotalTimeWatched' column to numeric minutes, group by 'DateID', and sum the total time watched for each date.



```python
def convert_to_minutes(time_str):
    parts = time_str.split(' ')
    hours = int(parts[0].replace('h', '')) if 'h' in parts[0] else 0
    minutes = int(parts[1].replace('m', '')) if len(parts) > 1 and 'm' in parts[1] else 0
    return hours * 60 + minutes

df['TotalTimeWatched_minutes'] = df['TotalTimeWatched'].apply(convert_to_minutes)

daily_viewership = df.groupby('DateID')['TotalTimeWatched_minutes'].sum().reset_index()
display(daily_viewership.head())
```

**Reasoning**:
Create a time series plot of the aggregated daily viewership data to visualize trends over time.



```python
plt.figure(figsize=(12, 6))
plt.plot(daily_viewership['DateID'], daily_viewership['TotalTimeWatched_minutes'])
plt.xlabel("Date")
plt.ylabel("Total Time Watched (minutes)")
plt.title("Daily Viewership Over Time")
plt.grid(True)
plt.show()
```
